## 라인 플롯 차트 그리기

1. 데이터 준비  
2. x축과 y축 데이터 지정 및 라인 플롯 생성: plt.plot(x,y)  
3. 차트 제목 설정: plt.title()  
4. x축 레이블 설정: plt.xlabel()    
5. y축 레이블 설정: plt.ylabel()  
6. 라인플롯 표시: plt.show()  

```python
# pyplot 모듈 임포트
import matplotlib.pyplot as plt

#1. 데이터 준비
x = range(2016,2024)
y = [350,410,520,695,543,720,695,777]

#2. 라인플롯 생성
plt.plot(x,y)

#3. 차트 제목 설정
plt.title('Annual Sales')

#4. x축 레이블 설정
plt.xlabel('years')

#5. y축 레이블 설정
plt.ylabel('sales')

#6. 라이플롯 표시
plt.show()
```

## 바 차트 그리기

1. 데이터 준비  
2. x축과 y축 데이터 지정 및 바 플롯 생성: plt.plot(x,y)  
3. 차트 제목 설정: plt.title()  
4. x축 레이블 설정: plt.xlabel()    
5. y축 레이블 설정: plt.ylabel()  
6. 눈금 이름 리스트 생성: xLabel = ['first','second','third','fourth']  
7. 바 차트의 x축 눈금 이름 설정: plt.xticks()
8. 범례 설정: plt.legend()  
9. 바 차트 표시: plt.show()

```python
# pyplot 모듈 임포트
import matplotlib.pyplot as plt

#1. 데이터 준비
y1 = [350, 410, 520, 695]
y2 = [200,250,385,350]
x = range(len(y1))

#2. 바 차트 생성
plt.bar(x,y1,width=0.7, color='blue')
plt.bar(x,y2,width=0.7, color='red', bottom=y1)

#3. 차트 제목 설정
plt.title('Quarterly Sales')

#4. x축 레이블 설정
plt.xlabel('Quarters')

#5. y축 레이블 설정
plt.ylabel('sales')

#6. 눈금 이름 리스트 생성
xLabel = ['first','second','third','fourth']

#7. 바 차트 x 눈금 이름 설정
plt.xticks(x,xLabel,fontsize=10)

#8. 범례 설정
plt.legend(['chairs','desks'])

#9. 바 차트 표시
plt.show()
```

```python
# 3-4. 수평선, 수직선 표시
import matplotlib.pyplot as plt

x = np.arange(0, 4, 0.5)

# x에 대한 함수 정의 y = x + 1 , y = x*x , y = -2*x
plt.plot(x, x + 1, 'bo')
plt.plot(x, x**2 - 4, 'g--')
plt.plot(x, -2*x + 3, 'r:')

# 수평선 : axhline , 수직선: axvline 
plt.axhline(4.0, 0.1, 0.9, color='lightgray', linestyle='--', linewidth=2) # y깂, xmin, xmax, color, linestyle, linewidth
plt.axvline(1.0, 0.2, 0.8, color='lightgray', linestyle='--', linewidth=2) # x값, ymin, ymax , color, linestyle, linewidth 

# 수평선: hlines, 수직선: vlines
plt.hlines(-0.62, 1.0, 2.5, color='lightgreen', linestyle='solid', linewidth=3) # y깂, xmin, xmax, color, linestyle, linewidth
plt.vlines(1.8, -3.0, 2.0, color='lightgreen', linestyle='solid', linewidth=3)  # x값, ymin, ymax , color, linestyle, linewidth 

plt.show()
```

```python
# 4 영역 Layout : 그리드 스펙 설정 
gs = GridSpec(3,3)
fig = plt.figure(figsize=(9,9))
# subplot 크기와 위치 생성
ax1 = fig.add_subplot(gs[0,:])
ax2 = fig.add_subplot(gs[1,:-1])
ax3 = fig.add_subplot(gs[1:,-1])
ax4 = fig.add_subplot(gs[2,0])
ax5 = fig.add_subplot(gs[2,1])

# 데이터 생성
x = np.linspace(0, 10, 100)
y = np.sin(x)
bar_labels = ['A', 'B', 'C', 'D']
bar_values = [3, 12, 5, 8]
scatter_x = np.random.randn(50)
scatter_y = np.random.randn(50)
pie_values = [15, 30, 45, 10]
pie_labels = ['Apple', 'Banana', 'Cherry', 'Date']


# 그래프 그리기
ax1.plot(x, y)
ax1.set_title('ax1: Line Graph')

ax2.bar(bar_labels, bar_values)
ax2.set_title('ax2: Bar Graph')

ax3.pie(pie_values, labels=pie_labels, autopct='%1.1f%%')
ax3.set_title('ax3: Pie Chart')

ax4.scatter(scatter_x, scatter_y)
ax4.set_title('ax4: Scatter Plot')

ax5.plot(x, np.cos(x))
ax5.set_title('ax5: Cosine Graph')

plt.show()
```

## Matplotlib patches 모듈을 사용하여 도형 시각화

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# 1. 기본 스타일 설정
plt.style.use('default')
plt.rcParams['figure.figsize'] = (5, 4)
plt.rcParams['font.size'] = 12

# 2. 사각형 정보를 나타내는 튜플의 리스트
rect = [(30, 60, 100, 200),
        (200, 30, 250, 450),
        (600, 430, 120, 170),
        (750, 120, 210, 300),
        (140, 550, 370, 220)]

# 3. add_patch()로 도형 (patches) 시각화하기
fig, ax = plt.subplots()

for (x, y, w, h) in rect:
    ax.add_patch(
        patches.Rectangle(
            (x, y), w, h,
            edgecolor='green',
            facecolor='palegreen',
            linewidth=0.5,
            fill=True))

# 4. 그래프 설정 및 띄우기
ax.set_xlim(0, 1000)
ax.set_ylim(0, 800)
plt.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go
%matplotlib inline
#주피터노트북에서 그래프 출력이 안될때   ## 주피터랩에서는 아직 지원이 안되고 있음 plotly 
import plotly.offline as pyo
# 오프라인 모드로 변경
pyo.init_notebook_mode()

# 2x2 서브플롯 생성, 원 그래프를 위한 서브플롯 유형 지정
fig = make_subplots(
    rows=2, cols=2,
    subplot_titles=("Line Plot", "Bar Chart", "Scatter Plot", "Pie Chart"),
    specs=[[{}, {}], [{}, {"type": "pie"}]]
)

# 첫 번째 서브플롯: 선 그래프
fig.add_trace(go.Scatter(x=[0, 1, 2, 3, 4], y=[0, 1, 4, 9, 16]), row=1, col=1)

# 두 번째 서브플롯: 막대 그래프
fig.add_trace(go.Bar(x=["A", "B", "C"], y=[1, 3, 2]), row=1, col=2)

# 세 번째 서브플롯: 산점도
fig.add_trace(go.Scatter(x=[0, 1, 2], y=[1, 3, 2], mode='markers'), row=2, col=1)

# 네 번째 서브플롯: 원 그래프
fig.add_trace(go.Pie(labels=["A", "B", "C"], values=[1, 3, 2]), row=2, col=2)

fig.show()
```

```python
from plotly.subplots import make_subplots
import plotly.graph_objects as go

# 2x2 서브플롯 생성, 원 그래프를 위한 서브플롯 유형 지정
# specs=[[{}, {}], [{}, {"type": "pie"}]]: 각 서브플롯의 유형 지정
# 첫 번째 행의 서브플롯은 디폴트 유형(일반적으로 XY 플롯)을 사용
# 두 번째 행의 두 번째 열에 위치한 서브플롯은 원 그래프("type": "pie")로 설정
fig = make_subplots(rows=2, cols=2,subplot_titles=("Line Plot", "Bar Chart", "Scatter Plot", "Pie Chart"),specs=[[{}, {}], [{}, {"type": "pie"}]])

#데이터 생성
x = np.linspace(0, 10, 100)
y = np.sin(x)
bar_labels = ['A', 'B', 'C', 'D']
bar_values = [3, 12, 5, 8]
scatter_x = np.random.randn(50)
scatter_y = np.random.randn(50)
pie_values = [15, 30, 45, 10]
pie_labels = ['Apple', 'Banana', 'Cherry', 'Date']


# 첫 번째 서브플롯: 선 그래프
fig.add_trace(go.Scatter(x=x, y=y), row=1, col=1)

# 두 번째 서브플롯: 막대 그래프
fig.add_trace(go.Bar(x=bar_labels, y=bar_values), row=1, col=2)

# 세 번째 서브플롯: 산점도
fig.add_trace(go.Scatter(x=scatter_x, y=scatter_y, mode='markers'), row=2, col=1)

# 네 번째 서브플롯: 원 그래프
fig.add_trace(go.Pie(labels=pie_labels, values=pie_values), row=2, col=2)

# 이미지 크기 설정
fig.update_layout(width=759, height=759)

fig.show()
# 출처: https://coduking.tistory.com/entry/Matplotlib-Subplot-활용방법
```
